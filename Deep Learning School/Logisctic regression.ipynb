# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
#делаем one-hot-encoding
dummy_features = pd.get_dummies(data[cat_cols],dtype='int',drop_first=True)
dummy_features
```

```python
#убираем столбик с целевой переменной
data_origin = data.drop(columns=['Churn'])
```

```python
data_origin = pd.concat([data_origin ,dummy_features], axis=1)
data_origin.drop(columns=cat_cols, inplace=True)
data_origin.head()
```

```python
#то же самое для тестовых данных
dummy_features_test = pd.get_dummies(data_test[cat_cols],dtype='int',drop_first=True)
data_test = pd.concat([data_test, dummy_features_test], axis=1)
data_test.drop(columns=cat_cols, inplace=True)
data_test.head()
```

```python
#делим данные
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_origin, data['Churn'].values,
                                                    train_size=0.8,
                                                    random_state=42)
```

```python
#нормируем признаки
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
data_test_scaled = scaler.transform(data_test)
```

```python
#используем LogisticRegressionCV для подбора гиперпараметров
reg = LogisticRegressionCV(cv=5, scoring='roc_auc', refit=True, max_iter=1000, random_state=42)
reg.fit(X_train_scaled, y_train)
y_test_pred = reg.predict_proba(X_test_scaled)[:, 1]
y_train_pred = reg.predict_proba(X_train_scaled)[:, 1]
```

```python
#оцениваем качество
from sklearn.metrics import roc_auc_score, roc_curve

train_auc = roc_auc_score(y_train, y_train_pred)
test_auc = roc_auc_score(y_test, y_test_pred)

plt.figure(figsize=(20,10))
plt.plot(*roc_curve(y_train, y_train_pred)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_pred)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
#Посмотрим, с каким коэффициентом получились такие результаты
print(f'Оптимальное значение C: {reg.C_[0]}')
```
